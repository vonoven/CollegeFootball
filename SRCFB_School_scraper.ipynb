```python
# import packages
import urllib3
import csv
from bs4 import BeautifulSoup
import re
import pandas as pd
import datetime

#settings
http = urllib3.PoolManager()
today = datetime.datetime.today()
schools = []

#scrape the data and write the .csv files
url = "https://www.sports-reference.com/cfb/schools/"
response = http.request('GET', url)
soup = BeautifulSoup(response.data)
cnt = 0
#print(soup)

for row in soup.findAll('tr'):
    try:
        col1=row.findAll('th')
        Rk=col1[0].string
        col=row.findAll('td')
        link=row.find('a')
        http_name=link.get('href')
        School = col[0].get_text()
        From = col[1].get_text()
        To = col[2].get_text()
        Yrs = col[3].get_text()
        Overall_G = col[4].get_text()
        Overall_W = col[5].get_text()
        Overall_L = col[6].get_text()
        Overall_T = col[7].get_text()
        Overall_Pct = col[8].get_text()
        Bowls_G = col[9].get_text()
        Bowls_W = col[10].get_text()
        Bowls_L = col[11].get_text()
        Bowls_T = col[12].get_text()
        Bowls_Pct = col[13].get_text()
        SRS = col[14].get_text()
        SOS = col[15].get_text()
        AP = col[16].get_text()
        CC = col[17].get_text()
        Notes = col[18].get_text()
        schools.append([School, http_name, From, To, Yrs, Overall_G, Overall_W, Overall_L, Overall_T, Overall_Pct,
                        Bowls_G, Bowls_W, Bowls_L, Bowls_T, Bowls_Pct, SRS, SOS, AP, CC, Notes])
        cnt += 1
    except:
        pass
print("Finished writing " + str(cnt) + " schools")
```

```python
df=pd.DataFrame(schools,columns=['School', 'HTTP', 'From', 'To', 'Yrs', 'Overall_G', 'Overall_W', 'Overall_L',
                                 'Overall_T', 'Overall_Pct', 'Bowls_G', 'Bowls_W', 'Bowls_L', 'Bowls_T',
                                 'Bowls_Pct', 'SRS', 'SOS', 'AP', 'CC', 'Notes'])
df
```

```python
df.to_csv('./SRinput/schools.csv')
```
