## This notebook was used to obtain a list of standard school names from the website, translate them into believed HTTP names and check for errors.  Should only ever need to be run again if new schools are introduced (or names change over time for some reason)

### Get list of valid school standardized names

```python
schools = []
url = "https://www.sports-reference.com/cfb/schools/"
response = http.request('GET', url)
soup = BeautifulSoup(response.data, features="html.parser")
cnt = 0
for row in soup.findAll('tr'):
    try:
        col1=row.findAll('th')
        Rank=col1[0].string
        col=row.findAll('td')
        school = col[0].get_text()
        
        schools.append(school)
        cnt += 1
    except:
        pass
print("Finished scraping with " + str(cnt) + " rows")
```

### Translate each name to the believed HTTP name

```python
httpschool = []
for school in schools:
    nw = school.lower().replace(' ', '-').replace('(', '').replace(')', '').replace('&', '').replace('--', '-').replace('\'', '')
    if nw == 'louisiana':
        nw = 'louisiana-lafayette'
    elif nw == 'lsu':
        nw = 'louisiana-state'
    elif nw == 'ole-miss':
        nw = 'mississippi'
    elif nw == 'pitt':
        nw = 'pittsburgh'
    elif nw == 'smu':
        nw = 'southern-methodist'
    elif nw == 'uab':
        nw = 'alabama-birmingham'
    elif nw == 'ucf':
        nw = 'central-florida'
    elif nw == 'usc':
        nw = 'southern-california'
    elif nw == 'utep':
        nw = 'texas-el-paso'
    elif nw == 'utsa':
        nw = 'texas-san-antonio'
    httpschool.append(nw)
```

### Hit the schools website and check if each one worked

```python
# import packages
import urllib3
import certifi
import csv
from bs4 import BeautifulSoup
import re
import pandas as pd
import datetime

#settings
http = urllib3.PoolManager(
        cert_reqs='CERT_REQUIRED',
        ca_certs=certifi.where())
naughty = []

for school in httpschool:
    #scrape the data and write the .csv files
    url = "https://www.sports-reference.com/cfb/schools/" + school + "/"
    response = http.request('GET', url)
    soup = BeautifulSoup(response.data, features="html.parser")
    #print(soup)

    for row in soup.findAll('h1'):
        data = row.get_text()
    if data == 'Page Not Found (404 error)':
        naughty.append(school)
    else:
        print(school)
```
