# Model Predictions

### XGBoost and Keras together

```python
from jobs import SRmodel
XG = SRmodel.predict_XGB()
KS = SRmodel.predict_Keras()
Pred_final = XG.copy()
Pred_final['Keras'] = KS['Prediction']
Pred_final['Keras_proba'] = KS['Proba']
Pred_final.columns = ['Gamecode', 'Date', 'Team', 'Opp', 'XG', 'Keras', 'Keras_proba']
Pred_final
```

```python
Pred_final['Confidence'] = Pred_final.apply(lambda x: abs(.5 - x['Keras_proba']), axis=1)
```

```python
Pred_final = Pred_final.sort_values(by=['Date'])
Pred_final
```

```python
Pred_final.to_csv('./SRoutput/weekly_preds/2018_bowl_preds_w_SOS.csv')
```

### Prediction - Easily readable version

```python
import pandas as pd

preds = pd.read_csv('./SRoutput/weekly_preds/2018_bowl_preds_w_SOS.csv')
cols = ['Date', 'Team', 'Opp', 'Keras', 'Keras_proba', 'Confidence']
preds = preds[cols]
preds.head()
```

```python
preds['Winner'] = preds.apply(lambda x: x['Team'] if int(x['Keras']) else x['Opp'], axis=1)
preds['Conf'] = preds.apply(lambda x: round(x['Confidence']*100, 2), axis=1)
```
