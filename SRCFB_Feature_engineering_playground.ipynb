```python
from jobs import SRgamelogs
import pandas as pd

past_games = pd.read_csv('./SRinput/current/past_games.csv')
past_games.set_index('Season', inplace=True)
past_games['Date'] = pd.to_datetime(past_games['Date'])
sum_stats = ['Pass_cmp', 'Pass_att', 'Pass_yds', 'Pass_TD', 'Rush_att', 'Rush_yds', 
             'Rush_TD', 'TO_plays', 'TO_yds', 'FD_pass', 'FD_rush', 'FD_pen', 
             'FD_tot', 'Pen_num', 'Pen_yds', 'Turn_fum', 'Turn_int', 'Turn_tot']
mean_stats = ['Pass_pct', 'Rush_avg', 'TO_avg']
for stat in sum_stats:
    past_games['Team_Off_' + stat] = past_games.apply(lambda x: 
                                        SRgamelogs.get_GL_team_sumstat_STD(stat, x['Team'], 
                                                                            x['Date'], x.name, 'O'), axis=1)
for stat in mean_stats:
    past_games['Team_Off_' + stat] = past_games.apply(lambda x: 
                                        SRgamelogs.get_GL_team_meanstat_STD(stat, x['Team'], 
                                                                            x['Date'], x.name, 'O'), axis=1)
for stat in sum_stats:
    past_games['Team_Def_' + stat] = past_games.apply(lambda x: 
                                        SRgamelogs.get_GL_team_sumstat_STD(stat, x['Team'], 
                                                                            x['Date'], x.name, 'D'), axis=1)
for stat in mean_stats:
    past_games['Team_Def_' + stat] = past_games.apply(lambda x: 
                                        SRgamelogs.get_GL_team_meanstat_STD(stat, x['Team'], 
                                                                            x['Date'], x.name, 'D'), axis=1)

past_games
```
